<!-- Area Plots -->

```python
import pandas as pd
import numpy as np
from numpy.random import randn
import matplotlib.pyplot as plt
```

```python
# creating our dataframe
df = pd.DataFrame( randn(25,4), columns=["Mon", "Tues", "Wed", "Thurs"] )
df
```

```python
# area plot
# stacked = False karna pada because of mix of negative and positive values.
df.plot(kind = 'area', stacked=False)
```

```python
# dataframe ki negative values ko positive kar diya mod laga ke.
df.abs().plot(kind = 'area')
```

```python
# alpha represent tranceparency
df.abs().plot(kind = 'area', alpha = 0.3)
```

```python
df.abs().plot(kind = 'area', title = "my plot")
```

```python
# legend false karne se mon,tue  wala sticker hat gya.
df.abs().plot(kind = 'area', title = "my plot", legend=False)
```

```python
# want to plot just one column
df["Mon"].abs().plot(kind = 'area', title = "my plot", legend=True)
```

```python
# Two columns
df[ ["Mon", "Wed"] ].abs().plot(kind = 'area', title = "my plot", legend=True)
```

```python
# the number of row of table equals number of columns in the dataframe.
df.abs().plot(kind = 'area', title = "my plot", legend=True, table=True)
```
