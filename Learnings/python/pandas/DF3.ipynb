<!-- Adding & Removing columns in dataframe -->

```python
import pandas as pd
import numpy as np
```

```python
"""
this is how a dataframe is included in vs code.

"""

# Replace 'anish.csv' with your CSV file path
# we can write 'anish.csv' wherever we have used csv_file_path variable.
file = 'data.csv'

try:
    # Try reading the CSV file with default encoding
    df = pd.read_csv(file)
except UnicodeDecodeError:
    # If there's an encoding error, try with a different encoding
    df = pd.read_csv(file, encoding='latin1')

# Display the DataFrame
print(df)
```

```python
# add column from a python list. 
# length of list should be equal to the number of rows.
# we can create a list on our own also to do so.
allone = np.ones((32,1), dtype='int64')         # ek random column banaya to add in the dataframe. df me 32 elements hai therefore 32 elements ki list bnai hai.
df["Random"] = allone
print(df)
```

```python
# giving default value throughout the column.
df["Alive/Dead"] = [True]*len(df)           # value "True" daali hai jo ki bool hai, multiply by length of df kiya taaki uski length ka column bane.
df["name"] = ["bob"]*len(df)
df
```

```python
# can store nan values
df["extra"] = [np.nan]*len(df)
df
```

```python
# add columns using insert() fun - can choose position because as default columns last me add hote hai.

# .insert(column_position, column_name, column_list, bool(do you want duplicates?) )
# if last me False hoga then list me duplicate value would not be allowed.
df.insert(1, "age", [5]*len(df), True)
df 
```

```python
# if you want to keep your original data frame as it is and want to create a column making a new data frame.
# new dataframe hi ban gya.
my_df = df.assign(Horse=[False]*len(df))
my_df
```

```python
# it does not have a column named horse.
df
```

```python
# removing a column
# axis=1 karna pada to specify our columns, it kinda takes you to the row of column_headings.
# axis=0 is the row headings, by default axis is equal to 0.
df.drop("Random", axis=1)
```

```python
# running the dataframe again, we see that random is still there
df
```

```python
df["Random"]
```

```python
# to delete it permanently, change the inplace to True.
df.drop("Random", axis=1, inplace=True)
```

```python
# random has been permanently deleted.
df
```

```python
# permanently deleting a column in very first place.
df.drop("name", axis=1, inplace=True)
```

```python
# name column also get deleted.
# these deletions have nothing to do with the .csv file, that will remain as it is.
df
```

```python
# remove a row
# the first row gets deleted.
# df.drop(1) bhi chalega but prefer the following.
# make inplace=True to delete it permanently.
df.drop(1, axis=0)      # good practice.
```
